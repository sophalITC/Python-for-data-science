## draw random samples from a normal distribution

```python
height=np.random.normal(170, 3, 1000) # mu=170, sd=3, n=1000
height
```

```python
np.mean(height)
```

```python
np.std(height)
```

```python
plt.hist(height)
```

## rock, paper, scissors

```python
g=np.array(['rock', 'paper', 'scissors'])
g
```

```python
np.random.choice(g)
```

```python
p1=np.random.choice(g)
p2=np.random.choice(g)
print(f'{p1}-{p2}')
```

```python
for _ in range(10):
    p1=np.random.choice(g)
    p2=np.random.choice(g)
    print(f'{p1}-{p2}')
```

```python
pet=np.array(['cat', 'dog', 'bird', 'fish'])
c=np.random.choice(pet, size=100, replace=True)
c
```

```python
u, counts=np.unique(c, return_counts=True)
print(u)
print(counts)
```

```python
plt.bar(u, counts)
```

```python
pet=np.array(['cat', 'dog', 'bird', 'fish'])
c=np.random.choice(pet, size=10000, replace=True, p=[.2, .4, .1, .3])
c
```

```python
u, counts=np.unique(c, return_counts=True)
print(u)
print(counts)
```

```python
plt.bar(u, counts)
```
