```python
import pandas as pd 
from IPython.display import HTML
```

```python
jason_url='https://rss.applemarketingtools.com/api/v2/kh/music/most-played/25/music-videos.json'
data=pd.read_json(jason_url)
```

```python
data
```

```python
pd.set_option('display.max_colwidth', -1)
```

```python
df=pd.DataFrame(data['feed']['results'])
df.head()
```

```python
df['artwork']=df.apply(lambda v: f'<img src={v.artworkUrl100} width=100, height=100/>',axis=1)
df['artwork']
```

```python
df['song_url']=df.apply(lambda v:f'<a href={v.url}>{v["name"]}</a>', axis=1)
df['song_url']
```

```python
HTML(df[['artistName','artwork','song_url']].to_html(escape=False))
```
