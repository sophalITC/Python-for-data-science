```python
import numpy as np
celcius=np.arange(0,101)
```

```python
#far=cel*1.8+32
farenheit=celcius*1.8+32
```

```python
t=np.stack((celcius,farenheit),axis=1)
t
```

```python
np.savetxt('temperature.csv',t)
```

```python
np.savetxt('temp2.csv',t,delimiter=',',fmt='%.1f')
```

```python
np.savetxt('temp3.csv',t,
          delimiter=',',
          fmt='%.2f',
          header='celcious,farenheit')
```

```python
np.savetxt('temp4.csv',t,
          delimiter=',',
          fmt='%.1f',
          header='celcius,farenheit',
          comments='')
```
