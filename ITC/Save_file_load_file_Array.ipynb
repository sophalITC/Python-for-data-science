### savetxt (save ndarray to text file)
### loadtxt (load data from text file to numpy array) 

```python
import numpy as np
```

```python
#array range (0 ==>100) celciuos 
#fah =cel*(9/5)+32  
cel=np.arange(101)
cel
```

```python
fah=cel*(9/5)+32
fah
```

```python
a=np.stack((cel,fah),axis=1)
a
```

```python
#csv 
np.savetxt('temperature.csv',a)
```

```python
np.savetxt('tem2.csv',a,
          delimiter=',',
          fmt='%.1f',
          header='celious,fahranheit')
```

```python
np.savetxt('tem3.csv',a,
          delimiter=',',
          fmt='%.1f',
          header='celious,fahranheit',
          comments=' ')
```

### load file

```python
import numpy as np
r=np.loadtxt('temperature.csv')
r
```

```python
r2=np.loadtxt('tem2.csv', delimiter=',')
r2
```

```python
r3=np.loadtxt('tem3.csv', delimiter=',',skiprows=1)
r3
```
