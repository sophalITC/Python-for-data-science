```python
import pandas as pd
%matplotlib inline
```

```python
df=pd.read_csv('https://github.com/sophalITC/datasci/raw/master/mpg.csv')
df.sample(6)
```

```python
df.plot.scatter(x='displ',y='cty');
```

```python
%config InlineBackend.figure_format = 'retina'
```

```python
df.plot.scatter(x='displ',y='cty');
```

```python
df.plot.scatter(x='displ',y='cty',grid=True,title=' City mile per gallon vs engine size');
```

```python
df.sample(4)
```

```python
df.plot.scatter(x='displ',y='cty',grid=True,
                c=df.drv.map({'4':'blue','f':'green','r':'red'}),
               title=' City mile per gallon vs engine size');
```

```python
df.plot.scatter(x='displ',y='cty',grid=True,
                c=df.drv.map({'4':'blue','f':'green','r':'red'}),
                s=df.cyl*15,
               title=' City mile per gallon vs engine size');
```

```python
df.plot.scatter(x='displ',y='cty',grid=True,
                c=df.drv.map({'4':'blue','f':'green','r':'red'}),
                s=df.cyl*15,
                alpha=.5,
                ylim=(0,40),
                xlim=(0,8),
               title=' City mile per gallon vs engine size');
```
